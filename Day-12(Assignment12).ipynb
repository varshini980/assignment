```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, output_image='wordcloud.png'):
    """
    Generates a WordCloud from the given text and saves it as an image.

    Args:
        text (str): The input text for the WordCloud.
        output_image (str): The file name to save the WordCloud image.

    Returns:
        None
    """
    # Create a WordCloud object
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)
    
    # Display the WordCloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
    # Save the WordCloud image
    wordcloud.to_file(output_image)
    print(f"WordCloud saved as '{output_image}'")

# Test the function with the provided text
text = 'data science machine learning artificial intelligence'
generate_wordcloud(text)
```
